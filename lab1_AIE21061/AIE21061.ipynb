```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

y,sr = librosa.load('speech processing.wav')
plt.figure(figsize=(10,5))
librosa.display.waveshow(y)
plt.title('Speech Signal Waveform')
plt.show()
```

```python
#A2

print("Length of the signal:",len(y),"samples")
print("Sampling rate:",sr, "Hz")
print("Magnitude range:",np.min(y),"to", np.max(y))
```

```python
#A3

import IPython.display as ipd

start = int(1 * sr)
end = int(2 * sr)
segment = y[start:end]
librosa.display.waveshow(segment)
ipd.Audio(segment, rate=sr)
```

```python
#A4


segment_1_start = int(1 * sr)
segment_1_end = int(2 * sr)
segment_2_start = int( 2* sr)
segment_2_end = int(3 * sr)
segment_1 = y[segment_1_start:segment_1_end]
segment_2 = y[segment_2_start:segment_2_end]
print("Playing Segment 1:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1)
plt.title('Segment-1')
plt.show()
ipd.Audio(segment_1, rate=sr)
```

```python
#A5

print("Playing Segment 2:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2)
plt.title('Segment-1')
plt.show()
ipd.Audio(segment_2, rate=sr)
```
